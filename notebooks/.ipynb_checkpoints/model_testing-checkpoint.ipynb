```python
import numpy as np
import torch
import os
import pickle
os.chdir('..')
from data_processing.generate_dataset import generate_spectrogram_2d, generate_spectrogram
from models import SpectraMLP
```

```python
load_path = 'model_checkpoints/best_model.pth'
n_sample = 1000

testset = None
concentration = None
for i in range(n_sample):
    x, spectrogram, params = generate_spectrogram(interpolate=True)
    if i == 0:
        testset = spectrogram
        concentration = params
    else:
        testset = torch.cat((testset,spectrogram),0)
        concentration = torch.cat((concentration,params),0)
        
mlp_model = torch.load(load_path, map_location=torch.device('cpu'))
with torch.no_grad():
    pred_params = mlp_model(testset)
    
# Average error over test set

#Average predicted concentration of each molecules 
average_pred_concentration = torch.mean(pred_params,0)

# Average actual concentration of each molecules 
average_concentration = torch.mean(concentration,0)

# Error
error = torch.mean(torch.square(concentration-pred_params), 0)
```

```python
import matplotlib.pyplot as plt

molecules = ["water", "fat", "hbo2", "hb", "cytoa_oxy", "cytoa_red"]
plt.plot(molecules,average_pred_concentration, label='Predicted average concentration')
plt.plot(molecules,average_concentration, label='Average concentration')
plt.ylabel('Concentration level')
plt.xlabel('Molecules')
plt.suptitle('Concentration of molecules')
plt.legend()
```
